## **What is an Inner Product?**  
- A **generalization of the dot product** to abstract spaces.  
- Provides tools to measure **angles, lengths, and projections** in these spaces.  
- Used in:  
  - Linear Algebra  
  - Functional Analysis  
  - Quantum Mechanics  

---

## **Definition of an Inner Product**  
An **inner product** on a vector space $V$ is a function:  
$$
\langle \cdot, \cdot \rangle : V \times V \to F  
$$  
where $F$ is the field of **real** or **complex numbers**.  

It satisfies the following **properties** for all $u, v, w \in V$ and scalar $c \in F$:

---

## **Properties of Inner Product**  

1. **Additivity in Both Arguments**:  
   $$
   \langle u + v, w \rangle = \langle u, w \rangle + \langle v, w \rangle  
   $$  
   for all $u, v, w \in V$.

2. **Homogeneity**:  
   $$
   \langle c u, v \rangle = c \langle u, v \rangle  
   $$  
   for any scalar $c$.

3. **Linearity in the First Argument**:  
   For scalars $a, b \in F$:  
   $$
   \langle au + bv, w \rangle = a \langle u, w \rangle + b \langle v, w \rangle  
   $$

4. **Conjugate-Linearity in the Second Argument** *(for complex fields)*:  
   $$
   \langle u, cv \rangle = \overline{c} \langle u, v \rangle  
   $$  
   where $\overline{c}$ is the **complex conjugate** of $c$.

5. **Conjugate Symmetry**:  
   $$
   \langle u, v \rangle = \overline{\langle v, u \rangle}  
   $$

6. **Positive Definiteness**:  
   $$
   \langle v, v \rangle \geq 0 \quad \text{and} \quad \langle v, v \rangle = 0 \iff v = 0  
   $$

---

# Examples
## 1. Dot Product in Euclidean Space

In $\mathbb{R}^n$, the inner product is the familiar dot product:

$$
\langle u, v \rangle = u \cdot v = u_1 v_1 + u_2 v_2 + \cdots + u_n v_n
$$

This satisfies all the inner product properties.

---

## 2. Complex Inner Product

For $\mathbb{C}^n$, the inner product includes complex conjugation:

$$
\langle u, v \rangle = \overline{u_1} v_1 + \overline{u_2} v_2 + \cdots + \overline{u_n} v_n
$$

---

## 3. Function Spaces

In a space of square-integrable functions $L^2$, the inner product is defined as:

$$
\langle f, g \rangle = \int_a^b f(x) \overline{g(x)} \, dx
$$

This is widely used in Fourier series and functional analysis.

## 4. Inner Product in Higher-Dimensional Spaces

In higher-dimensional spaces, the inner product can be defined using **kernels** in Support Vector Machines (SVM), such as:

$$
K(x, y) = \langle \phi(x), \phi(y) \rangle
$$

where $\phi(x)$ and $\phi(y)$ are the mappings of $x$ and $y$ into higher-dimensional feature spaces, and $K(x, y)$ represents the kernel function that computes the inner product in this transformed space.


---

# Applications of Inner Product

1. **Length (Norm)**: The length of a vector $v$ is given by:

$$
\| v \| = \langle v, v \rangle^{1/2}
$$

2. **Angle Between Vectors**: The angle $\theta$ between two vectors $u$ and $v$ is defined as:

$$
\cos \theta = \frac{\langle u, v \rangle}{\| u \| \| v \|}
$$

3. **Orthogonality**: Two vectors $u$ and $v$ are orthogonal if:

$$
\langle u, v \rangle = 0
$$

4. **Projections**: The projection of a vector $v$ onto $u$ is:

$$
\text{proj}_u(v) = \frac{\langle v, u \rangle}{\langle u, u \rangle} u
$$

# Inner Product in Dual Form of Soft-Margin SVM

## Primal Form of Soft-Margin SVM Objective Function

In the **primal form** of the soft-margin SVM, the objective function is:

$$
\min_{w, b, \xi} \frac{1}{2} \| w \|^2 + C \sum_{i=1}^n \xi_i
$$

**Subject to:**

$$
y_i ( w \cdot x_i + b ) \geq 1 - \xi_i \quad \text{for all} \quad i
$$

$$
\xi_i \geq 0 \quad \text{for all} \quad i
$$

where $\xi_i$ are the slack variables, which measure the degree of misclassification, and $C$ is the regularization parameter that controls the trade-off between the margin size and misclassification.

---

## Dual Form of Soft-Margin SVM

The **dual form** of the soft-margin SVM introduces Lagrange multipliers $\alpha_i$ and involves the inner product:

$$
\max_{\alpha} \sum_{i=1}^n \alpha_i - \frac{1}{2} \sum_{i=1}^n \sum_{j=1}^n \alpha_i \alpha_j y_i y_j \langle x_i, x_j \rangle
$$

**Subject to the constraints:**

$$
0 \leq \alpha_i \leq C \quad \text{for all} \quad i
$$

$$
\sum_{i=1}^n \alpha_i y_i = 0
$$

Here, $\langle x_i, x_j \rangle$ represents the inner product (or kernel) between feature vectors $x_i$ and $x_j$.


## Relationship Between Kernel Trick and Inner Product

## Inner Product Limitation:
In the dual form of SVM, the decision boundary is linear in the input space. But real-world data often requires non-linear boundaries.

## The Kernel Trick:
Replace 
$$
\langle x_i, x_j \rangle
$$ 
with a kernel function 
$$
K(x_i, x_j).
$$
$K(x_i, x_j)$ implicitly computes the inner product in a higher-dimensional feature space:
$$
K(x_i, x_j) = \langle \phi(x_i), \phi(x_j) \rangle
$$
No need to explicitly transform $x$ to $\phi(x)$.

## Example Kernels:
- **Linear**: 
$$
K(x, y) = x \cdot y
$$
(no transformation).
- **Polynomial**:
$$
K(x, y) = (x \cdot y + 1)^d
$$
- **RBF (Gaussian)**: 
$$
K(x, y) = \exp(-\gamma \|x - y\|^2)
$$

## Example: Polynomial Kernel with Degree 2

Consider two points in 2D: $x_1 = (1, 2)$ and $x_2 = (3, 4)$.

### 1: **Feature Mapping for Polynomial Kernel**

For the quadratic polynomial kernel, the feature mapping $\phi(x)$ for a 2D point $x = (x_1, x_2)$ with degree 2 should be:

$$
\phi(x) = (1, x_1, x_2, x_1^2, \sqrt{2} x_1 x_2, x_2^2)
$$

This ensures that both the linear and quadratic terms are properly captured, including the constant 1.

- For $x_1 = (1, 2)$:
  $$
  \phi(x_1) = (1, 1, 2, 1^2, \sqrt{2} \cdot 1 \cdot 2, 2^2) = (1, 1, 2, 1, 2\sqrt{2}, 4)
  $$

- For $x_2 = (3, 4)$:
  $$
  \phi(x_2) = (1, 3, 4, 3^2, \sqrt{2} \cdot 3 \cdot 4, 4^2) = (1, 3, 4, 9, 6\sqrt{2}, 16)
  $$

### 2: **Kernel Calculation**

The polynomial kernel between $x_1$ and $x_2$ is:

$$
K(x_1, x_2) = (x_1 \cdot x_2 + 1)^2
$$

The dot product between $x_1$ and $x_2$ is:

$$
x_1 \cdot x_2 = 1 \cdot 3 + 2 \cdot 4 = 3 + 8 = 11
$$

Now, compute the kernel:

$$
K(x_1, x_2) = (11 + 1)^2 = 12^2 = 144
$$

### 3: **Inner Product Calculation in Explicit Feature Space**

Now, let's compute the inner product in the transformed feature space:

$$
\langle \phi(x_1), \phi(x_2) \rangle = 1 \cdot 1 + 1 \cdot 3 + 2 \cdot 4 + 1 \cdot 9 + (2\sqrt{2}) \cdot (6\sqrt{2}) + 4 \cdot 16
$$

Simplify each term:

$$
\langle \phi(x_1), \phi(x_2) \rangle = 1 + 3 + 8 + 9 + 2 \cdot 6 \cdot 2 + 64
$$

$$
\langle \phi(x_1), \phi(x_2) \rangle = 1 + 3 + 8 + 9 + 24 + 64 = 144
$$

### Conclusion:

Both the explicit feature mapping inner product and the kernel trick yield the same result:

$$
\langle \phi(x_1), \phi(x_2) \rangle = K(x_1, x_2) = 144
$$

By adding the constant term **1** in the feature mapping $\phi(x)$ and using the polynomial kernel $(x \cdot y + 1)^2$, we can ensure that both methods give the same result.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

plt.style.use('dark_background')

X, y = make_moons(n_samples=300, noise=0.2, random_state=47)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=47
)

kernels = ['linear', 'poly', 'rbf']  
C_values = [0.1, 1, 10] 

def plot_decision_boundary(model, X, y, ax, title):
    h = 0.02  # Step size in the mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    
    # Meshgrid for plotting decision surface
    xx, yy = np.meshgrid(
        np.arange(x_min, x_max, h), np.arange(y_min, y_max, h)
    )
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # Plot decision boundary and test data points
    ax.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm_r)
    ax.scatter(X[y == 0, 0], X[y == 0, 1], color='magenta', 
               edgecolors='k', label="Negative class")
    ax.scatter(X[y == 1, 0], X[y == 1, 1], color='cyan', 
               edgecolors='k', label="Positive class")
    ax.set_xlabel("Feature 1")
    ax.set_ylabel("Feature 2")
    ax.set_title(title)
    ax.legend()


fig, axes = plt.subplots(len(C_values), len(kernels), 
                         figsize=(18, 12), dpi=300)

for i, C in enumerate(C_values):
    for j, kernel in enumerate(kernels):
        if kernel == 'poly':  
            model = SVC(kernel=kernel, degree=2, C=C)
        else:
            model = SVC(kernel=kernel, C=C)
        
        # Train the model and predict on the test set
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        
        # Plot the decision boundary for the test set
        plot_decision_boundary(
            model, X_test, y_test, axes[i, j], 
            f"{kernel.capitalize()} Kernel (C={C})\nAccuracy: {accuracy:.2f}"
        )

# Adjust layout
plt.tight_layout()
plt.savefig('output.jpg', dpi=300)
plt.show()
```
